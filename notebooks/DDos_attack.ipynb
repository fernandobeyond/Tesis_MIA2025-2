```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from sklearn.utils.class_weight import compute_class_weight
```

```python
assert "Attack_Label" in df.columns, "No encuentro Attack_Label en df"
df["Attack_Label"] = df["Attack_Label"].astype(int)

X = df.drop(columns=["Attack_Label"])
y = df["Attack_Label"]
```

```python

num_cols = [c for c in X.columns if np.issubdtype(X[c].dtype, np.number)]
cat_cols = [c for c in X.columns if c not in num_cols]

print(f"Total filas: {len(df)}, num_cols={len(num_cols)}, cat_cols={len(cat_cols)}")
```

```python
TEST_SIZE = 0.20
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, stratify=y, random_state=42
)
print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
numeric_pipe = Pipeline([('imp', SimpleImputer(strategy='median')),
                         ('sc', StandardScaler())])
categorical_pipe = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                             ('oh',  OneHotEncoder(handle_unknown='ignore'))])

preproc = ColumnTransformer([('num', numeric_pipe, num_cols),
                             ('cat', categorical_pipe, cat_cols)])

Xtr_t = preproc.fit_transform(X_train)
Xte_t = preproc.transform(X_test)

# Si el train balanceado es 50/50, spw≈1; lo calculamos igual por robustez
pos = int((y_train==1).sum()); neg = int((y_train==0).sum())
scale_pos_weight = float(neg/pos) if pos>0 else 1.0

xgb_clf = xgb.XGBClassifier(
    n_estimators=600, max_depth=6, learning_rate=0.08,
    subsample=0.9, colsample_bytree=0.9, reg_lambda=1.0,
    random_state=42, n_jobs=-1, eval_metric='aucpr',
    scale_pos_weight=scale_pos_weight
)
xgb_clf.fit(Xtr_t, y_train, eval_set=[(Xte_t, y_test)], verbose=False)
p_xgb_test = xgb_clf.predict_proba(Xte_t)[:,1]

print("XGB  -> PR-AUC(test):", round(average_precision_score(y_test, p_xgb_test),4),
      "| ROC-AUC(test):", round(roc_auc_score(y_test, p_xgb_test),4))
```

```python
num_imp_sc = Pipeline([('imp', SimpleImputer(strategy='median')),
                       ('sc', StandardScaler())])
Xtr_num = num_imp_sc.fit_transform(X_train[num_cols])
Xte_num = num_imp_sc.transform(X_test[num_cols])

WINDOW, STEP = 20, 1  # puedes ajustar
def build_sequences(Xmat, yvec, window=WINDOW, step=STEP):
    Xs, ys = [], []
    n = len(yvec)
    for i in range(0, n-window, step):
        Xs.append(Xmat[i:i+window])        # (window, n_feats)
        ys.append(yvec.iloc[i+window-1])   # etiqueta del último paso
    return np.array(Xs), np.array(ys)

Xtr_seq, ytr_seq = build_sequences(Xtr_num, y_train)
Xte_seq, yte_seq = build_sequences(Xte_num, y_test)
print("Seq shapes -> train:", Xtr_seq.shape, " test:", Xte_seq.shape)
```

```python
# Modelo CNN1D + LSTM (arquitectura base del paper)
inp = layers.Input(shape=(WINDOW, Xtr_seq.shape[-1]))
x = layers.Conv1D(64, 3, padding='same', activation='relu')(inp)
x = layers.MaxPooling1D(2)(x)
x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
x = layers.Dropout(0.3)(x)
x = layers.LSTM(128)(x)
x = layers.Dropout(0.3)(x)
emb = layers.Dense(64, activation='relu', name='deep_embedding')(x)
out = layers.Dense(1, activation='sigmoid')(emb)
cnn_lstm = models.Model(inp, out)
cnn_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
```

```python
# class weights para desbalance (por si no quedó 50/50)
classes = np.unique(ytr_seq)
cw = compute_class_weight(class_weight='balanced', classes=classes, y=ytr_seq)
class_weight = {int(c): float(w) for c,w in zip(classes, cw)}

es = callbacks.EarlyStopping(patience=5, restore_best_weights=True, monitor='val_auc', mode='max')
hist = cnn_lstm.fit(
    Xtr_seq, ytr_seq, validation_split=0.1,  # mini holdout interno solo para early stopping
    epochs=30, batch_size=256, class_weight=class_weight,
    callbacks=[es], verbose=1
)

p_cnn_test = cnn_lstm.predict(Xte_seq, batch_size=512).ravel()
print("CNN-LSTM -> PR-AUC(test, seq):", round(average_precision_score(yte_seq, p_cnn_test),4),
      "| ROC-AUC(test, seq):", round(roc_auc_score(yte_seq, p_cnn_test),4))
```

```python
# Para comparar con las secuencias: usamos la "cola" del vector tabular
p_xgb_test_al = p_xgb_test[-len(p_cnn_test):]
y_test_al     = np.array(y_test)[-len(p_cnn_test):]

p_blend_test = 0.5 * p_xgb_test_al + 0.5 * p_cnn_test  # promedio simple

print("BLEND -> PR-AUC(test):", round(average_precision_score(y_test_al, p_blend_test),4),
      "| ROC-AUC(test):", round(roc_auc_score(y_test_al, p_blend_test),4))

# Reporte final @0.5
y_pred = (p_blend_test >= 0.5).astype(int)
print("\n== Clasificación (test, blend @0.5) ==")
print(classification_report(y_test_al, y_pred, digits=4))
print("Matriz de confusión:\n", confusion_matrix(y_test_al, y_pred))

# (Opcional) Guardar modelos
import joblib
joblib.dump(xgb_clf, "model_xgb.joblib")
cnn_lstm.save("model_cnnlstm.keras")
print("\nModelos guardados: model_xgb.joblib, model_cnnlstm.keras")
```

```python
# ============================================================
# K-FOLD EVALUATION (k=5) — Híbrido XGB + CNN-LSTM + Blend
# ============================================================

import numpy as np, pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, average_precision_score,
    matthews_corrcoef, log_loss, brier_score_loss
)
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from sklearn.utils.class_weight import compute_class_weight

# ---------- Preparación (si hace falta) ----------
assert "Attack_Label" in df.columns, "No encuentro Attack_Label en df"
df["Attack_Label"] = df["Attack_Label"].astype(int)
X_all = df.drop(columns=["Attack_Label"])
y_all = df["Attack_Label"].values

# Identificar columnas numéricas/categóricas si no existen
try:
    num_cols
    cat_cols
except NameError:
    num_cols = [c for c in X_all.columns if np.issubdtype(X_all[c].dtype, np.number)]
    cat_cols = [c for c in X_all.columns if c not in num_cols]

# ---------- Configuración ----------
k = 5
WINDOW, STEP = 20, 1

# Preprocesamiento tabular para XGB
numeric_pipe = Pipeline([('imp', SimpleImputer(strategy='median')),
                         ('sc', StandardScaler())])
categorical_pipe = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                             ('oh',  OneHotEncoder(handle_unknown='ignore'))])
preproc = ColumnTransformer([('num', numeric_pipe, num_cols),
                             ('cat', categorical_pipe, cat_cols)])

def make_xgb(spw=1.0):
    return xgb.XGBClassifier(
        n_estimators=400, max_depth=6, learning_rate=0.08,
        subsample=0.9, colsample_bytree=0.9, reg_lambda=1.0,
        random_state=42, n_jobs=-1, eval_metric='aucpr',
        scale_pos_weight=spw
    )

def build_sequences(Xmat, yvec, window=WINDOW, step=STEP):
    Xs, ys = [], []
    n = len(yvec)
    for i in range(0, n-window, step):
        Xs.append(Xmat[i:i+window])
        ys.append(yvec[i+window-1])
    return np.array(Xs), np.array(ys)

def make_cnn_lstm(input_shape):
    inp = layers.Input(shape=input_shape)
    x = layers.Conv1D(64, 3, padding='same', activation='relu')(inp)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    x = layers.LSTM(128)(x)
    x = layers.Dropout(0.3)(x)
    emb = layers.Dense(64, activation='relu')(x)
    out = layers.Dense(1, activation='sigmoid')(emb)
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

rows = []
cm_sum = np.zeros((2,2), dtype=int)

fold = 0
for tr_idx, te_idx in skf.split(X_all, y_all):
    fold += 1
    print(f"\n=== Fold {fold}/{k} ===")

    X_tr, X_te = X_all.iloc[tr_idx], X_all.iloc[te_idx]
    y_tr, y_te = y_all[tr_idx], y_all[te_idx]

    # ----- Rama XGB (tabular)
    Xtr_t = preproc.fit_transform(X_tr)
    Xte_t = preproc.transform(X_te)
    pos = int((y_tr==1).sum()); neg = int((y_tr==0).sum())
    spw = float(neg/pos) if pos>0 else 1.0
    xgb_clf = make_xgb(spw)
    xgb_clf.fit(Xtr_t, y_tr, eval_set=[(Xte_t, y_te)], verbose=False)
    p_xgb = xgb_clf.predict_proba(Xte_t)[:,1]

    # ----- Rama CNN-LSTM (secuencias sobre NUMÉRICAS)
    # Escalado/imputación numérica SOLO con train del fold
    num_imp_sc = Pipeline([('imp', SimpleImputer(strategy='median')),
                           ('sc', StandardScaler())])
    Xtr_num = num_imp_sc.fit_transform(X_tr[num_cols])
    Xte_num = num_imp_sc.transform(X_te[num_cols])

    # Construir secuencias
    Xtr_seq, ytr_seq = build_sequences(Xtr_num, y_tr, WINDOW, STEP)
    Xte_seq, yte_seq = build_sequences(Xte_num, y_te, WINDOW, STEP)

    # Si por el window no hay suficientes ejemplos, saltar fold
    if len(ytr_seq) < 100 or len(yte_seq) < 100:
        print("Fold con muy pocas secuencias; se omite evaluación secuencial en este fold.")
        p_cnn = np.zeros_like(p_xgb[-len(p_xgb):])  # fallback
        y_eval = y_te[-len(p_cnn):]
        p_xgb_eval = p_xgb[-len(p_cnn):]
    else:
        cnn = make_cnn_lstm((WINDOW, Xtr_seq.shape[-1]))
        classes = np.unique(ytr_seq)
        cw = compute_class_weight(class_weight='balanced', classes=classes, y=ytr_seq)
        class_weight = {int(c): float(w) for c, w in zip(classes, cw)}

        es = callbacks.EarlyStopping(patience=5, restore_best_weights=True, monitor='val_auc', mode='max')
        cnn.fit(Xtr_seq, ytr_seq, validation_split=0.1,
                epochs=25, batch_size=256, class_weight=class_weight,
                callbacks=[es], verbose=0)

        p_cnn = cnn.predict(Xte_seq, batch_size=512).ravel()

        # Alinear “cola” del vector tabular al largo secuencial
        p_xgb_eval = p_xgb[-len(p_cnn):]
        y_eval     = yte_seq  # ya alineado por construcción

    # ----- Blend (promedio simple)
    p_blend = 0.5 * p_xgb_eval + 0.5 * p_cnn
    y_pred  = (p_blend >= 0.5).astype(int)

    # ----- Métricas del fold
    cm = confusion_matrix(y_eval, y_pred, labels=[0,1])
    tn, fp, fn, tp = cm.ravel()
    specificity = tn/(tn+fp) if (tn+fp)>0 else 0.0
    npv = tn/(tn+fn) if (tn+fn)>0 else 0.0

    row = {
        "fold": fold,
        "support": int(len(y_eval)),
        "accuracy": accuracy_score(y_eval, y_pred),
        "precision": precision_score(y_eval, y_pred, zero_division=0),
        "recall": recall_score(y_eval, y_pred, zero_division=0),
        "f1": f1_score(y_eval, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_eval, p_blend),
        "pr_auc": average_precision_score(y_eval, p_blend),
        "specificity": specificity,
        "npv": npv,
        "mcc": matthews_corrcoef(y_eval, y_pred),
        "log_loss": log_loss(y_eval, p_blend, eps=1e-15),
        "brier": brier_score_loss(y_eval, p_blend),
        "tn": int(tn), "fp": int(fp), "fn": int(fn), "tp": int(tp)
    }
    rows.append(row)
    cm_sum += cm

# ----- Resultados por fold y resumen
cv_df = pd.DataFrame(rows)
display(cv_df)

summary = pd.DataFrame({
    "metric": ["accuracy","precision","recall","f1","roc_auc","pr_auc","specificity","npv","mcc","log_loss","brier"],
    "mean":   [cv_df[m].mean() for m in ["accuracy","precision","recall","f1","roc_auc","pr_auc","specificity","npv","mcc","log_loss","brier"]],
    "std":    [cv_df[m].std(ddof=1) for m in ["accuracy","precision","recall","f1","roc_auc","pr_auc","specificity","npv","mcc","log_loss","brier"]],
})
print("\n=== Promedio ± std (k-fold, híbrido blend) ===")
display(summary)

print("\n=== Matriz de confusión agregada (sumada sobre folds, blend) ===")
cm_agg = pd.DataFrame(cm_sum, index=["true_0","true_1"], columns=["pred_0","pred_1"])
display(cm_agg)
```
